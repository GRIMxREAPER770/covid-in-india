```python
# --------------------------------------------
# Beginner-Friendly Complete Data Analysis
# Dataset: District Wise Rainfall
# --------------------------------------------

# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load the dataset
file_path = "/content/district wise rainfall normal.csv"
data = pd.read_csv(file_path)

# 3. View first few rows
print("------ FIRST 5 ROWS ------")
print(data.head())

# 4. Basic information about dataset
print("\n------ DATASET INFO ------")
print(data.info())

# 5. Check for missing values
print("\n------ MISSING VALUES ------")
print(data.isnull().sum())

# 6. Summary statistics of numeric columns
print("\n------ SUMMARY STATISTICS ------")
print(data.describe())

# 7. Check column names
print("\n------ COLUMN NAMES ------")
print(list(data.columns))

# -----------------------------
# ðŸ“Š DATA ANALYSIS & VISUALS
# -----------------------------

# A) Number of districts per state
district_counts = data['STATE_UT_NAME'].value_counts()
plt.figure(figsize=(12,6))
district_counts.plot(kind="bar", color="skyblue")
plt.title("Number of Districts per State/UT")
plt.xlabel("State/UT")
plt.ylabel("District Count")
plt.xticks(rotation=90)
plt.show()

# B) Average annual rainfall by state
plt.figure(figsize=(12,6))
state_avg_rain = data.groupby("STATE_UT_NAME")["ANNUAL"].mean().sort_values(ascending=False)
state_avg_rain.plot(kind="bar", color="green")
plt.title("Average Annual Rainfall by State/UT")
plt.xlabel("State/UT")
plt.ylabel("Average Rainfall (mm)")
plt.xticks(rotation=90)
plt.show()

# C) Seasonal rainfall correlation heatmap
# Automatically detect seasonal columns (anything except identifiers)
possible_columns = [col for col in data.columns if col not in ["STATE_UT_NAME", "DISTRICT"]]
print("\nSeasonal/Rainfall Columns considered:", possible_columns)

plt.figure(figsize=(8,6))
sns.heatmap(data[possible_columns].corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation between Rainfall Parameters")
plt.show()

# D) Compare top 10 wettest districts (by Annual Rainfall)
plt.figure(figsize=(12,6))
top10 = data.groupby("DISTRICT")["ANNUAL"].mean().sort_values(ascending=False).head(10)
top10.plot(kind="bar", color="orange")
plt.title("Top 10 Wettest Districts (Annual Rainfall)")
plt.ylabel("Annual Rainfall (mm)")
plt.show()

# E) Boxplot of rainfall distribution across states
plt.figure(figsize=(12,6))
sns.boxplot(x="STATE_UT_NAME", y="ANNUAL", data=data)
plt.title("Rainfall Distribution by State/UT")
plt.xticks(rotation=90)
plt.show()

print("\nâœ… Complete Beginner-Friendly Analysis Completed!")
```
